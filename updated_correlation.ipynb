# Correlation

## Strategy

At each time step, for each cryptocurrency in our universe:
  - Calculate the returns of the cryptocurrency
  - Shift the returns forward in time
  - Over the last N periods, for each coin A:
      - Determine the other coin B with the highest correlation of returns of A to returns shifted forward one period of B
      - Calculate the OLS regression of shifted forward returns of B on to returns of A
   - For the next K periods, calculate the expected return of each coin A based on the regression from forward returns of its correlated coin B
   - If the predicted return of A is more than its actual return for a period, assign A a numeric signal to short it in the next period in proportion to this difference. Otherwise assign A a numeric signal to long it in proportion to this difference.
